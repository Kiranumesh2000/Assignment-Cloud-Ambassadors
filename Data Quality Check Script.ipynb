```python
# lets treat outliers using IQR method
def detect_outlier(x):
    # taking 5,25,75 percentile of column
    q25=np.percentile(x,25)
    q75=np.percentile(x,75)
    #calculationg IQR range
    IQR=q75-q25
    #Calculating minimum threshold
    lower_bound=q25-(1.5*IQR)
    upper_bound=q75+(1.5*IQR)
    #Capping outliers
    return upper_bound,lower_bound
```

```python
 uvw,lvw = detect_outlier(Data["bmi"])
Data["bmi"] = np.where(Data["bmi"] > uvw, uvw, Data["bmi"])
Data["bmi"] = np.where(Data["bmi"] < lvw, lvw, Data["bmi"])
```

```python
 uvw,lvw = detect_outlier(Data["daily_avg_steps"])
Data["daily_avg_steps"] = np.where(Data["daily_avg_steps"] > uvw, uvw, Data["daily_avg_steps"])
Data["daily_avg_steps"] = np.where(Data["daily_avg_steps"] < lvw, lvw, Data["daily_avg_steps"])
```

```python
#As all the outliers are treated print the boxplot again
sns.boxplot(Data["bmi"])
```
